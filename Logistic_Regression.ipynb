```r
for (col in names(fraud)) {
  if (is.numeric(fraud[[col]])) {
    fraud[[col]][is.na(fraud[[col]])] <- mean(fraud[[col]], na.rm = TRUE)
  }
}
```

```r
set.seed(123) # for reproducibility
train_index <- sample(1:nrow(fraud), 0.8 * nrow(fraud))
train_data <- fraud[train_index, ]
test_data <- fraud[-train_index, ]
```

```r
logistic_model <- glm(Class ~ ., data = train_data, family = "binomial")
```

```r
predictions <- predict(logistic_model, newdata = test_data, type = "response")
```

```r
predicted_classes <- ifelse(predictions > 0.5, 1, 0)
```

```r
confusion_matrix <- confusionMatrix(factor(predicted_classes), factor(test_data$Class), mode = "prec_recall")
```

```r
confusion_matrix
```

```r
accuracy <- mean(predicted_classes== test_data$Class)
```

```r
print(paste("Accuracy:", accuracy))
```

```r
predicted_probabilities <- predict(logistic_model, newdata = test_data, type = "response")
```

```r
predicted_classes <- ifelse(predicted_probabilities > 0.5, 1, 0)
```

```r
rmse <- sqrt(mean((predicted_probabilities - test_data$Class)^2))
print(paste("RMSE:", rmse))
```

```r
install.packages("ROCR")
```

```r
library(ROCR)
```

```r
prediction_obj <- prediction(predicted_probabilities, test_data$Class)
```

```r
perf <- performance(prediction_obj, "tpr", "fpr")
precision_recall <- performance(prediction_obj, "prec", "rec")
```

```r
plot(perf, colorize = TRUE, main = "ROC Curve", col.main = "blue", lwd = 2)
```

```r
plot(precision_recall, colorize = TRUE, main = "Precision-Recall Curve", col.main = "blue", lwd = 2)
```
